```python
age_price = age_purchase_df.groupby("Purchase Count").mean()["Price"]
age_value = age_purchase_df.groupby("Purchase Count").sum()["Price"]
#age_list = age_purchase_df.groupby("SN").count()["SN"]
age_total = age_value/age_purchase
age_analysis = pd.DataFrame({
    "Purchase Count":age_purchase,
    "Average Purchase Price":age_price,
    "Total Purchase Value":age_value,
    "Avg Total Purchase per Person":age_total
})
age_analysis
```
